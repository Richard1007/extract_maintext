## Get URL

```python
doi_html = '/Users/richardpang/Desktop/extract_maintext-1/JMR doi html/JMR2000-2005.html'
doi_lst = []
with open(doi_html, 'r') as f:
    soup = BeautifulSoup(f, 'html.parser')
titles=soup.find_all('dl',{'class':"print-citation"})
for i in range(len(titles)):
    
    if "DOI:" in titles[i].get_text():
        doi = titles[i].get_text().split('DOI: ')[1].split(' ')[0][:-1]
        # print(doi)
    # print(doi)
        doi_lst.append(doi)
df=pd.DataFrame(doi_lst)
len(df)
```

# Scraping

```python
#For ChromeDriver version 79.0.3945.16 or over
options = Options()
options.add_argument('--disable-blink-features=AutomationControlled') 

#Open Browser
driver = webdriver.Chrome(executable_path='/Users/richardpang/.wdm/drivers/chromedriver/mac64/106.0.5249/chromedriver',options=options)
article_lst = []
count = 0
for i in range(len(url_lst)):
# driver = webdriver.Chrome( '/Users/richardpang/.wdm/drivers/chromedriver/mac64/106.0.5249/chromedriver')
    driver.get(url_lst[i])
    count += 1
    if count%10 ==0:
        print(count, 'articles have been collected.')
    element = driver.find_elements(By.TAG_NAME, 'div')
    element = driver.find_elements(By.CLASS_NAME, 'core-container')
    # print(len(element))
    if len(element)>4:
        title = element[0].text.split('\n')[3]
        try:
            abstract = element[2].text.split('\n')[1]
        except IndexError:
            continue 
        try:
            main_text = element[3].text
        except IndexError:
            continue 
        doi = doi_lst[i]
        article  = [title,abstract,main_text,doi]
        article_lst.append(article)
    
        
    # sleep(np.random.randint(6, 7))
    sleep(6)
driver.quit()
article_df = pd.DataFrame(article_lst)
article_df
```
